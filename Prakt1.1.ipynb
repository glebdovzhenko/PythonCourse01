## Работа со списками

```python
# создание списков
# просто перечисляем элементы
l1 = [1, 7, 4, 3, 2]
print('l1 = ', l1)
# генерируем
l2 = [x for x in range(0, 5)]  # заполняем список целыми числами с 0 до 5, левая граница включается, правая исключается
print('l2 = ', l2)
l3 = [x for x in range(10, 23, 3)]  # заполняем список числами с 10 до 23 с шагом 3
print('l3 = ', l3)
# можно создавать списки из других списков
l4 = [x * 2 for x in l3]
print('l4 = ', l4)
```

```python
# Доступ к элементам списков производится по индексам.
# Индекс может быть положительным 0 <= ii < len(l1)
print('l1 = ', l1)
print('Нулевой элемент:', l1[0], 'Первый элемент:', l1[1], 'Второй элемент:', l1[2])
# Если индекс отрицательный, отсчёт идёт с конца
print('Последний элемент:', l1[-1], 'Предпоследний элемент:', l1[-2])
# С помощью индексов можно делать выборки (slices)
# С первого элемента по четвёртый (левая граница включительно, правая исключительно)
print(l1[1:4])
# Синтаксис слайсов:
# l1[start:stop:step]
# Причём можно опускать один или два аргумента. 
# Если аргумент опущен, он заменяется значением по умолчанию: start==0, stop==-1, step==1
# Следующие варианты записи эквивалентны
print(l1[0:2:1], l1[:2], l1[0:2],l1[:2:1],l1[:2:])
```

```python
# Удаление элементов из списка по индексу
l5 = ['q', 'w', 'e', 'r', 't', 'y']
print(l5)
del l5[0]  # удаляем нулевой элемент
print(l5)
# удаляем элементы с 2 по 4
del l5[2:4]
print(l5)
# удаляем элементы по индексу:
l5.pop(1)
print(l5)
```

```python
# удаляем элементы по значению (удаляется первый слева элемент с заданным значением)
l5 = ['q', 'w', 'e', 'r', 't', 'y']
print(l5)
l5.remove('e')
print(l5)
```

```python
# Поиск индексов элементов в списке
l5 = ['q', 'w', 'e', 'r', 't', 'y']
print(l5)
idx_e = l5.index('e')
print('Индекс элемента \'e\':', idx_e)
```

```python
# объединение списков
l6 = [2, 5, 4]
print(l5 + l6)
```

## Итерации и циклы

```python
l1 = ['a', 's', 'd', 'f', 'g', 'h']

# просто печатаем список
print(l1)

# перебираем элементы списка
# на каждой итерации в переменную el записывается один из элементов списка
for el in l1:
    print(el)
```

```python
# Перебирая элементы списка его нельзя изменить! можно, например, создать новый из выбранных элементов
l2 = [1, 3, 4, 5, 7, 2, 9, 5]
l2_even = []  # создаём пустой список
for el in l2:  # перебираем элементы l2 записывая их в переменную el
    if el % 2 == 0:  # для каждого элемента проверяем равен ли остаток от деления на 2 нулю
        l2_even.append(el)  # если остаток от деления на 2 равен 0, добавляем этот элемент в l2_even
print('l2 =', l2)
print('l2_even =', l2_even)
```

```python
# изменять список можно только обращаясь к его элементам по индексам
# Пример с перебором индексов:
l2 = [1, 3, 4, 5, 7, 2, 9, 5]
for ii in range(len(l2)):  # перебираем числа от 0 до длины списка l2 (исключая len(l2)) и записываем их в переменную ii
    print('Индекс', ii, 'Элемент', l2[ii])
```

```python
# Пример с удалением элементов:
l2 = [1, 3, 4, 5, 7, 2, 9, 5]
for ii in range(len(l2) - 1, 0, -1):  # перебираем числа от (len(l2)-1) до 0 с шагом -1, т.е. все индексы списка в обратном порядке: 7, 6, 5, ...
    print('Индекс', ii, 'Элемент', l2[ii])
    if l2[ii] % 2 == 0:
        print('Чётный, удаляем')
        l2.pop(ii)
    else:
        print('Нечётный, оставляем')
print('l2 после удаления элементов', l2)
```

### Почему индексы надо удалять в обратном порядке
Пусть у нас есть список `['a', 's', 'd', 'f', 'g', 'h']` и из него надо удалить элементы с индексами 3, 5.

Если сначала удалить элемент с индексом 3, произойдёт следующее:

`['a', 's', 'd', 'f', 'g', 'h'] -> ['a', 's', 'd', 'g', 'h']`

Видно, что у второго списка элементы с четвертого и далее сместились влево, и элемента с индексом 5 уже нет.

При этом, если сначала удалить пятый элемент, а потом третий, случится следующее:

`['a', 's', 'd', 'f', 'g', 'h'] -> ['a', 's', 'd', 'f', 'g'] -> ['a', 's', 'd', 'g']`

```python
# Пример с заменой элементов:
l2 = [1, 3, 4, 5, 7, 2, 9, 5]
print(l2)
for ii in range(len(l2)):  # перебираем числа от (len(l2)-1) до 0 с шагом -1, т.е. все индексы списка в обратном порядке: 7, 6, 5, ...
    if l2[ii] == 3:
        l2[ii] = 'три'  # записываем в список l2 в ячейку с индексом ii строку 'три'
    if l2[ii] == 7:
        l2[ii] = 'семь'
print(l2)
```

## Работа со словарями

```python
# Заполняем словарь из двух списков
# список ключей
ks = [1, 2, 3]
# список значений
vs = 'a', 'b', 'c'
# создаём пустой словарь
result = dict()

for ii in range(len(ks)):  # перебираем индексы с 0 до длины списка
    result[ks[ii]] = vs[ii]  # заполняем словарь парами ключ-значение 
print(result)
```
