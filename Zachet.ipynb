# Задание на зачёт: обработка экспериментальных данных СИ

В папке `Zachet` находится набор дифрактограмм снятых в процессе in situ эксперимента на станции СИ и лог-файл. 
В процессе эксперимента происходил нагрев образца. Температуры соответствующие каждой дифрактограмме отражены в лог-файле.
На зачёт нужно прочесть дифрактограммы, сделать фит пика, построить график зависимости положения пика от температуры.
Имейте ввиду, что этот набор файлов специально уменьшен. 
Пишите программу так, чтобы количество файлов (которых в реальности было ~400) можно было изменить, и программа продолжила работать.

### Задание:
Есть дифракционный пик, находящийся в диапазоне $2\Theta \in [3.55^{\circ}, 3.7^{\circ}] $. Сделать его фит распределением Гаусса, построить значение положения пика в $2\Theta$ от температуры. Температура записана в 7 колонке лог-файла (если считать с 0).

### Подсказки:
Функции, которые могут быть полезны:
`scipy.optimize.curve_fit`, `np.genfromtxt`, `os.listdir`, `filter`, `str.endswith`
